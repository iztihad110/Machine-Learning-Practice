```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

arr = np.random.rand(100, 5)

# print(arr)

df = pd.DataFrame(arr, columns = ['A', 'B', 'C', 'D', 'E'])

print(df)

col_mean = df.mean()
col_mean.plot(kind = "bar", color = "red", edgecolor = "black")
plt.title("Median")
plt.show()

col_median = df.median()
col_median.plot(kind = "bar", color = "green", edgecolor = "black")
plt.title("Median")
plt.show()

col_var = df.var()
col_var.plot(kind = "bar", color = "blue", edgecolor = "black")
plt.title("Median")
plt.show()

col_std = df.std()
col_std.plot(kind = "bar", color = "pink", edgecolor = "black")
plt.title("Median")
plt.show()
```

```python
text = "the dog and the cat chased the cat but the dog ran away"

words = text.split()

word_counts = pd.Series(words).value_counts()

word_counts.head(10).plot(kind = "bar", color = "orange", edgecolor = "black")

plt.show()

```

```python
words = ["the", "dog", "and", "the", "cat", "chased", "the", "cat", "but", "the", "dog", "ran", "away"]

stopwords = {"the", "and", "but"}

filtered_words = [w for w in words if not w in stopwords]

res =  pd.Series(filtered_words).value_counts()

res.plot(kind = "bar", color = "red", edgecolor = "black")

```

```python
df_corr = pd.DataFrame(np.random.randn(100, 5), columns=list("ABCDE"))
corr_matrix = df_corr.corr()
plt.imshow(corr_matrix, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```

```python
sentences = ["Hello Word", "She sells sea shells at the sea shore", "Life is Nothing"]

lengths = [ len(s.split()) for s in sentences]

df = pd.DataFrame({"Sentences": sentences, "Lengths": lengths})

print(df)

df["Lengths"].plot(kind = "hist", bins = 2, color = "violet", edgecolor = "black")
```

```python
matrix = np.random.randint(1, 10, size = (3, 3))

print(matrix)

matrix_det = np.linalg.det(matrix)
print(matrix_det)

matrix_inv = np.linalg.inv(matrix)
print(matrix_inv)

eig_vals, eig_vecs = np.linalg.eig(matrix)

print(f"Eigen Values{eig_vals}")
print(eig_vecs)
```
