```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y, alpha, epochs):
    w_curr = b_curr = 0

    m = len(x)

    plt.scatter(x, y, color = "red")

    for epoch in range(epochs):
        y_pred = w_curr * x + b_curr

        plt.plot(x, y_pred, color = "green")

        pd_w = ((1/m) * sum(x * (y_pred - y)))
        pd_b = ((1/m) * sum(y_pred - y))

        w_curr = w_curr - alpha * pd_w
        b_curr = b_curr - alpha * pd_b

```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 6, 8, 14])

gradient_descent(x, y, 0.01, 100000)
```
